```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import os
```

```python
data=pd.read_csv('/data/input/titanic/train.csv')

data.shape
print(data.head)
```

```python
data.describe()
```

```python
data.isna().sum()
```

```python
#visualising the missing values
msno.matrix(data)
```

```python
#data cleaning
#replacing the missing and negative age values with the median
#removing the the cabin cloumn as lot of missing data
data.Age=data.Age.fillna(0)
data.loc[data["Age"] < 0, "Age"] = data["Age"].median()
data.loc[data["Age"]==0,"Age"]=data["Age"].median()

data.drop('Cabin',axis=1,inplace=True)

```

```python
#encoding
label_encoder = LabelEncoder()

data['Sex'] = label_encoder.fit_transform(data['Sex'])
```

```python
data
```

```python
data.drop("PassengerId", axis=1, inplace=True)
data.drop("Name", axis=1, inplace=True)
data.drop("Ticket", axis=1, inplace=True)
data.drop("Embarked", axis=1, inplace=True)
data
```

```python
#splitting the data into test and train data

X=data.iloc[:,1:]
y=data.iloc[:,0]
print(X,y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#1.logistic Regression-simpler and fastest

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_logreg_pred = logreg.predict(X_test)
accpercent = logreg.score(X_test,y_test)*100
print(accpercent)
```

The accuracy of logistic regression model is 80.447%

```python
#2.KNN model

knn=KNeighborsClassifier()
knn.fit(X_train,y_train)
y_knn_pred=knn.predict(X_test)
acc_knn_percent=knn.score(X_test,y_test)*100
print(acc_knn_percent)
```

The accuracy of the knn model is 73.74%

```python
#3.Decision tree
dectree=DecisionTreeClassifier()
dectree.fit(X_train,y_train)
y_dct_pred=dectree.predict(X_test)
acc_dct_percent=dectree.score(X_test,y_test)*100
print(acc_dct_percent)
```

The accuracy of the decision tree model is 79.88%
